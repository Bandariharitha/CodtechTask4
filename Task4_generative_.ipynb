```python
# Import libraries
from transformers import GPT2LMHeadModel, GPT2Tokenizer
import torch
```

```python
# Load pre-trained GPT-2 model and tokenizer
model_name = "gpt2"  # You can also try 'gpt2-medium', 'gpt2-large', etc.
tokenizer = GPT2Tokenizer.from_pretrained(model_name)
model = GPT2LMHeadModel.from_pretrained(model_name)
model.eval()
```

```python
# Define text generation function
def generate_text(prompt, max_length=200, temperature=0.7, top_k=50):
    input_ids = tokenizer.encode(prompt, return_tensors='pt')

    with torch.no_grad():
        output = model.generate(
            input_ids,
            max_length=max_length,
            temperature=temperature,
            top_k=top_k,
            do_sample=True,
            num_return_sequences=1,
            pad_token_id=tokenizer.eos_token_id
        )

    generated_text = tokenizer.decode(output[0], skip_special_tokens=True)
    return generated_text
```

```python
# Try user prompt
user_prompt = "In 2050, climate change will have caused"
output = generate_text(user_prompt)
print("📝 Generated Text:\n")
print(output)
```
